# ## Tryng for R, using the empirical equation
# ###  $R_{NS} = \frac{\dot{E}^{1/6}}{B_{s}^{1/3}P^{2/3}}$

```python
R = (((o['Edot'])**(1/6))/((o['B_s'])**(1/3))*((o['po'])**(2/3)))
o['Radius']=R
```

Inertia

```python
k = o['Edot']
l = o['po']
m = o['pdot']
MI = (k*l**3)/(4*((3.14)**2)*m)
o['I']=MI
```

# ### From I , B, P and Pdot

```python
R2 = ((o['I']*o['pdot']*o['po'])/(o['B_s']**2))**(1/6)
o['R_2']=R2
```
